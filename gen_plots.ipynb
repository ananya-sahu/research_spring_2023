```python
import pandas as pd 
from matplotlib import pyplot as plt

df_all = pd.read_csv("./change_point_all_sims_scalars.csv")
df_emotions = pd.read_csv("./change_point_all_emotions_scalars.csv")
df_musc = pd.read_csv("/home/as5957/research_spring_2023/change_point_musc_sims_scalars.csv")
df_pose = pd.read_csv("/home/as5957/research_spring_2023/change_point_pose_sims_scalars.csv")
```

```python
df_all = df_all.dropna()
df_emotions = df_emotions.dropna()
df_musc = df_musc.dropna()
df_pose = df_pose.dropna()
```

```python
#seperate based on time_scalar
def seperate_scalar(df, feature):
    
    df_low = df[df['scalars'] > 3]

    df_high = df[df['scalars'] < 3]

    df_med = df[df['scalars'] == 3]

    y_low = [df_low['sim1'].mean(),df_low['sim2'].mean(),df_low['sim3'].mean()]
    y_high = [df_high['sim1'].mean(),df_high['sim2'].mean(),df_high['sim3'].mean()]
    y_med = [df_med['sim1'].mean(),df_med['sim2'].mean(),df_med['sim3'].mean()]

    x = ['frame-5','frame_t','frame+5']

    ig, ax = plt.subplots()
    ax.plot(x, y_low)
    ax.plot(x, y_high)
    ax.plot(x, y_med)
    legend_drawn_flag = True
    plt.legend(["scalar < 3", "scalar > 3", "scalar == 3"], loc=0, frameon=legend_drawn_flag)
    plt.xlabel("frames")
    plt.ylabel("similarity")
    title = "similarity in " + feature +" seperated by impact scalar"
    plt.title(title)
    plt.show()


seperate_scalar(df_all, "all")
```

```python
seperate_scalar(df_emotions, "emotions")
```

```python
seperate_scalar(df_musc, "facial movement")
```

```python
seperate_scalar(df_pose, "pose")
```
